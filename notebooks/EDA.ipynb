```python
import os
import cv2
import matplotlib.pyplot as plt

base = r"D:\Developments\Resume projects\brain-tumor-project\data\processed"

# Count images per split/class
for split in ["train", "val", "test"]:
    for cls in ["yes", "no"]:
        count = len(os.listdir(os.path.join(base, split, cls)))
        print(f"{split}/{cls}: {count} images")

# Show sample images
sample_path = os.path.join(base, "train", "yes")
sample_img = cv2.imread(os.path.join(sample_path, os.listdir(sample_path)[0]))
plt.imshow(cv2.cvtColor(sample_img, cv2.COLOR_BGR2RGB))
plt.title("Sample Tumor Image (train/yes)")
plt.axis("off")
plt.show()
```
